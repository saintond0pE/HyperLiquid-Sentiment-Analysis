```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
sns.set(style="whitegrid")
```

```python
sentiment_url = "https://drive.google.com/uc?id=1PgQC0tO8XN-wqkNyghWc_-mnrYv_nhSf"
trades_url = "https://drive.google.com/uc?id=1IAfLZwu6rJzyWKgBToqwSmmVYU6VbjVs"

df_sentiment = pd.read_csv(sentiment_url)
df_trades = pd.read_csv(trades_url)

print("Sentiment Shape:", df_sentiment.shape)
print("Trades Shape:", df_trades.shape)
```

```python
df_sentiment.head()
df_trades.head()
```

```python
print("\n--- SENTIMENT DATA ---")
print(df_sentiment.info())
print(df_sentiment.isnull().sum())
print("Duplicates:", df_sentiment.duplicated().sum())

print("\n--- TRADES DATA ---")
print(df_trades.info())
print(df_trades.isnull().sum())
print("Duplicates:", df_trades.duplicated().sum())
```

```python
# clean column names (VERY IMPORTANT)
df_sentiment.columns = df_sentiment.columns.str.strip().str.lower()
df_trades.columns = df_trades.columns.str.strip().str.lower().str.replace(' ', '_')

print(df_sentiment.columns)
print(df_trades.columns)
```

```python
# sentiment date
df_sentiment['date'] = pd.to_datetime(df_sentiment['date'], errors='coerce').dt.date

# trades timestamp
df_trades['timestamp_ist'] = pd.to_datetime(df_trades['timestamp_ist'], errors='coerce')
df_trades['date'] = df_trades['timestamp_ist'].dt.date

# drop invalid
df_trades = df_trades.dropna(subset=['date'])
df_sentiment = df_sentiment.dropna(subset=['date'])
```

```python
df = df_trades.merge(
    df_sentiment[['date', 'classification']],
    on='date',
    how='left'
)

df.head()
```

```python
df['closed_pnl'] = pd.to_numeric(df['closed_pnl'], errors='coerce')
df['size_usd'] = pd.to_numeric(df['size_usd'], errors='coerce')

# leverage not present → skip for now

df['win'] = df['closed_pnl'] > 0
df['abs_size'] = df['size_usd'].abs()
```

```python
daily_metrics = df.groupby(['date', 'classification']).agg({
    'closed_pnl': 'sum',
    'win': 'mean',
    'account': 'nunique',
    'size_usd': 'count',
    'abs_size': 'mean'
}).rename(columns={
    'closed_pnl': 'total_pnl',
    'win': 'win_rate',
    'account': 'unique_traders',
    'size_usd': 'num_trades',
    'abs_size': 'avg_trade_size'
}).reset_index()

daily_metrics.head()
```
